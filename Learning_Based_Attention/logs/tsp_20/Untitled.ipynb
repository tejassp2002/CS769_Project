```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from math import exp, log
```

```python
# %matplotlib widget
names = ["run-tsp20_rollout_2_20220506T052233-tag-val_avg_reward.csv",
"run-tsp20_rollout_3_20220506T055611-tag-val_avg_reward.csv",
"run-tsp20_rollout_4_20220506T062146-tag-val_avg_reward.csv"]
fig = plt.figure()
labels = ["2","3","4"]
for i in range(3):
    x,y = [],[] 
# opening the CSV file
    with open(names[i], mode ='r')as file:
   
      # reading the CSV file
      csvFile = csv.reader(file)
 
      # displaying the contents of the CSV file
      for lines in csvFile:
            if lines[0]=="Wall time":
                pass
            else:
                x.append(float(lines[1]))
                y.append(float(lines[2]))
            
    plt.plot(x,y,label=labels[i])
    
leg = plt.legend()
plt.xlabel("step", fontsize=12)
plt.ylabel("Avg. Validation Reward", fontsize=12)  
# # tp = [1e-3]*len(x)
# # plt.plot(x,tp)
# # tp2 = [5e-3]*len(x)
# # plt.plot(x,tp2)
plt.savefig("plots/layerreward.pdf")
plt.show()
```
